```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

vectorizer = TfidfVectorizer()
transform = vectorizer.fit_transform(new_data['description'])
cs = cosine_similarity(transform, transform)
cs
```

```python
data = new_data['title']
```

```python
def netflix_recom(title, cs = cs):
  title_search = pd.Index(data).get_loc(title)
  sim = list(enumerate(cs[title_search]))
  sim = sorted(sim, key = lambda x:x[1], reverse = True)
  top_10 =sim[1:11]
  recom = [i[0] for i in top_10]
  return new_data[['title', 'description']].iloc[recom]
```
